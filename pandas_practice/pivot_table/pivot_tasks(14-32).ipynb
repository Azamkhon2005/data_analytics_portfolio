```python
#8
'''Write a Pandas program to create a Pivot table and count survival by gender,
categories wise age of various classes'''
age_group = pd.cut(data['age'],[0,10,30,60,80])
data.pivot_table(index=['sex',age_group],columns='class',values='survived',aggfunc='count')
```

```python
#10
'''Write a Pandas program to create a Pivot table and find survival rate by gender,
age of the different categories of various classes. Add the fare as a dimension of columns
and partition fare column into 2 categories based on the values present in fare columns. '''
fare = pd.cut(data['fare'],3)
data.pivot_table(index=['sex',age_group],columns=['class',fare],values='survived')
```
